```python
#standartization
import numpy as np
import pandas as pd
from scipy import stats
iris = pd.read_csv(r"C:\Users\Fuat\Learning Machine Learning\Iris.csv")
```

```python
iris.head(3)
```

```python

LengthCm = iris['SepalLengthCm']
meanSepalLength = np.mean(LengthCm)
stdSepalLength = np.std(LengthCm)
X= (np.arange(150, dtype=float))
print(X)

for i in range(150):
    X[i] = ((LengthCm[i]-meanSepalLength)/stdSepalLength)
```

```python
X
```

```python
# normalization
maxLength = np.max(LengthCm)
minLength = np.min(LengthCm)
Y= (np.arange(150, dtype=float))
print(Y)
for i in range(150):
    Y[i] = ((Y[i]- minLength)/(maxLength - minLength))
```
