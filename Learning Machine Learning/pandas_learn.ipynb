```python
#Using iris dataset
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
iris = fetch_ucirepo(id=53) 
  
# data (as pandas dataframes) 
X = iris.data.features 
y = iris.data.targets 
  
# metadata 
print(iris.metadata) 
  
# variable information 
print(iris.variables) 
```

```python
# Convert X and y to a single DataFrame for easier manipulation
iris_df = pd.concat([X, y], axis=1)
```

```python
# Display the first 5 rows of the dataset
print(iris_df.head())
```

```python
# Display basic information about the dataset
print(iris_df.info())
```

```python
# Display summary statistics for the numerical columns
print(iris_df.describe())
```

```python
# Check for missing values
print(iris_df.isnull().sum())
```

```python
print(X.columns)  # Check the feature column names
print(y.columns)  # Check the target column name
```

```python
# Rename the target column to 'species'
iris_df = iris_df.rename(columns={'class': 'species'})

# Check the column names again
print(iris_df.columns)
# Count the number of samples for each species
print(iris_df['species'].value_counts())
```
